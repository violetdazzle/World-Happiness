```python
import tkinter as tk
from tkinter import ttk
import pandas as pd
import ipynb.fs.full.World_Happiness_2024 as World_Happiness_2024
```

```python
root = tk.Tk()
root.title("World Happiness Report")
root.geometry("700x500") 
```

```python
df = World_Happiness_2024.df
df = df.reset_index()
```

```python
# sort countries alphabetically
countries = sorted(df['Country name'].tolist())
```

```python
# function to determine color based on score
def get_color(score):
    if score <= 5:
        red = int(255)
        green = int((score / 5) * 255)
    else:
        red = int((10 - score) / 5 * 255)
        green = int(255)
    return f'#{red:02x}{green:02x}00'
```

```python
# function to transform score from 0-3 scale to 1-10 scale
def transform_score(mean_score):
    return 3 * mean_score + 1
```

```python
# defining the retrieve happiness score function
def retrieve_happiness_score():
    selected_country = clicked.get()
    copy_df = df.set_index('Country name')
    country_row = copy_df.loc[selected_country]
    mean_happiness_score = country_row['Mean Happiness Score']
    transformed_score = transform_score(mean_happiness_score)
    cluster_label = country_row['Cluster_Labels']
    
    # clear previous labels and canvas
    for widget in root.winfo_children():
        if isinstance(widget, tk.Label) or isinstance(widget, tk.Canvas):
            widget.destroy()
    
    # display Happiness Score
    happiness_label = tk.Label(root, text=f'Happiness Score for {selected_country}: {transformed_score:.2f}')
    happiness_label.pack(pady=10)
    
    # create a canvas to draw the slider with colored sections
    canvas = tk.Canvas(root, width=700, height=70)  # Increased width of the canvas
    canvas.pack(pady=10)
    
    # draw colored sections and labels on the canvas
    section_width = 700 / 10 
    for i in range(10):
        color = get_color(i + 1)
        canvas.create_rectangle(i * section_width, 0, (i + 1) * section_width, 50, fill=color, outline="")
        canvas.create_text((i + 0.5) * section_width, 60, text=str(i + 1), font=("Helvetica", 10, "bold"))
    
    # draw the slider pointer
    slider_position = (transformed_score - 1) * section_width
    slider_size = 15 
    canvas.create_oval(slider_position - slider_size, 20, slider_position + slider_size, 50, fill="white", width=2)
    
    # display cluster type
    if mean_happiness_score >= 1.5:
        cluster_type_label = tk.Label(root, text='Cluster: Happy Country', fg='green')
    else:
        cluster_type_label = tk.Label(root, text='Cluster: Sad Country', fg='red')
    cluster_type_label.pack(pady=10)
```

```python
# setup dropdown menu
clicked = tk.StringVar(root)
clicked.set('Choose a country')

country_dropdown = ttk.OptionMenu(root, clicked, *countries)
country_dropdown.pack(pady=10)
```

```python
# retrieve button
retrieve_button = ttk.Button(root, text="Retrieve Happiness Score", command=retrieve_happiness_score)
retrieve_button.pack(pady=10)
```

```python
# start tkinter event loop
root.mainloop()
```
